```python
import pandas as pd
from pytz import timezone

# load csv file
df = pd.read_csv('/Users/angelogentileiii/code/point-focal/russell/PointFocalFile-04_24.csv')
print(df.head())
```

```python
# Convert the 'Time' column from GMT to EST
df['Transaction Time (EST)'] = pd.to_datetime(df['Transaction Time (GMT)']).dt.tz_localize('GMT').dt.tz_convert('US/Eastern').dt.tz_localize(None)

# Sort the DataFrame by 'Transaction Time (EST)' in descending order
df = df.sort_values('Transaction Time (EST)', ascending=True)

# Define a function to categorize each trade into one of the three intervals
def get_interval(time):
    if time.hour < 9 or time.hour >= 16:
        return 'non-us-transaction'
    elif time.hour == 9 and time.minute < 30:
        return 'non-us-transaction'
    elif (time.hour == 9 and time.minute >= 30) or (time.hour == 10 and time.minute <= 30):
        return '9:30 - 10:30'
    elif time.hour >= 10 and time.hour < 15:
        return '10:31 - 14:59'
    else:
        return '15:00 - 16:00'

# Apply the function to the 'Transaction Time (GMT)' column to create a new 'Transection Interval' column
df['Transaction Time (EST)'] = pd.to_datetime(df['Transaction Time (EST)'])
df['Transaction Interval'] = df['Transaction Time (EST)'].apply(get_interval)
df  = df[df['Base Currency'] == 'USD']
df

```

```python
# Define the intervals
interval1 = '9:30 - 10:30'
interval2 = '10:31 - 14:59'
interval3 = '15:00 - 16:00'

# Filter the DataFrame for each interval
df_interval1 = df[df['Transaction Interval'] == interval1]
df_interval2 = df[df['Transaction Interval'] == interval2]
df_interval3 = df[df['Transaction Interval'] == interval3]

# Get the number of rows for each interval
num_rows_interval1 = df_interval1.shape[0]
num_rows_interval2 = df_interval2.shape[0]
num_rows_interval3 = df_interval3.shape[0]

print(num_rows_interval1+num_rows_interval2+num_rows_interval3)
print(f"Number of rows in interval 1: {num_rows_interval1}")
print(f"Number of rows in interval 2: {num_rows_interval2}")
print(f"Number of rows in interval 3: {num_rows_interval3}")
```

```python
# Calculate the percentage of trades in each transaction interval for entire basket

# Group the data by 'Order Id' and 'Transaction Interval', and calculate the total number of trades in each group
grouped_by_interval = df.groupby(['Transaction Interval']).size()

# Calculate the total number of trades for each Order Id' and for the entire basket
total_basket_trades = df.shape[0]

# Calculate the percentage of trades in each transection interval for each 'Order Id' and for the entire basket
percentage_by_interval = grouped_by_interval / total_basket_trades

percentage_by_interval_df = percentage_by_interval.to_frame()
percentage_by_interval_df.columns = ['Activity Percentage']
percentage_by_interval_df
```

```python
# Calculate the percentage of trades in each transaction interval for each 'Ticker'
grouped_by_ticker_interval = df.groupby(['Ticker', 'Transaction Interval']).size()
total_trades_by_ticker_interval = df.groupby('Ticker').size()
grouped_percentage_by_ticker= grouped_by_ticker_interval / total_trades_by_ticker_interval

grouped_percentage_by_ticker_interval_df = grouped_percentage_by_ticker.to_frame()
grouped_percentage_by_ticker_interval_df.columns = ['Activity Percentage']

grouped_percentage_by_ticker_interval_df
```

```python
# Calculate the percentage of trades in each transaction interval for each 'Ticker'
grouped_by_orderId = df.groupby(['Order Id', 'Transaction Interval']).size()
total_trades_by_orderId = df.groupby('Order Id').size()
grouped_percentage_by_orderId= grouped_by_orderId / total_trades_by_orderId

grouped_percentage_by_orderId_df = grouped_percentage_by_orderId.to_frame()
grouped_percentage_by_orderId_df.columns = ['Activity Percentage']

grouped_percentage_by_orderId_df
```

```python
# Calculate the percentage of trades by Ticker in each transaction interval for entire basket

percentage_by_ticker_interval_basket = grouped_by_ticker_interval / total_basket_trades

percentage_by_ticker_interval_basket_df = percentage_by_ticker_interval_basket.to_frame()
percentage_by_ticker_interval_basket_df.columns = ['Activity Percentage']
percentage_by_ticker_interval_basket_df
```

```python
# Calculate the percentage of trades by Order Id in each transaction interval for entire basket

percentage_by_orderId_interval_basket = grouped_by_orderId / total_basket_trades

percentage_by_orderId_interval_basket_df = percentage_by_orderId_interval_basket.to_frame()
percentage_by_orderId_interval_basket_df.columns = ['Activity Percentage']
percentage_by_orderId_interval_basket_df
```

```python
# Add the IEX OHLC close data from day prior
iex_close = pd.read_csv('/Users/angelogentileiii/code/point-focal/russell/IEX_OHLC_CLOSE.csv')
iex_close.rename(columns={'close': 'Close', 'symbol': 'Ticker'}, inplace=True)

# Aggregate unique tickers
unique_tickers = df.groupby('Ticker').agg({'Shares': 'sum', 'Side': 'first'}).reset_index()
unique_tickers.rename(columns={'Shares': 'Shares Total'}, inplace=True)

# Calculate total value and total shares per day
daily_total_val = df.groupby('Ticker').apply(lambda group: (group['Price'] * group['Shares']).sum()).reset_index(name='Total Value')

# Calculate total value and total shares (Interval)
interval_agg = df.groupby(['Ticker', 'Transaction Interval']).agg({'Shares': 'sum', 'Price': lambda x: (x * df.loc[x.index, 'Shares']).sum()}).reset_index()
interval_agg.rename(columns={'Shares': 'Shares Interval', 'Price': 'Interval Value'}, inplace=True)

# Merge unique_tickers with daily_total_val
calc_impshortfall_df = pd.merge(unique_tickers, daily_total_val, how='left', on='Ticker')

# Merge interval aggregates
calc_impshortfall_df = pd.merge(calc_impshortfall_df, interval_agg, how='left', on='Ticker')

# Merge with iex_close to get closing prices
calc_impshortfall_df = pd.merge(calc_impshortfall_df, iex_close[['Ticker', 'Close']], how='left', on='Ticker')

# Calculate VWAP per Day and VWAP per Interval
calc_impshortfall_df['VWAP Day'] = calc_impshortfall_df['Total Value'] / calc_impshortfall_df['Shares Total']  # Shares_x is total shares for the day
calc_impshortfall_df['VWAP Interval'] = calc_impshortfall_df['Interval Value'] / calc_impshortfall_df['Shares Interval']  # Shares_y is total shares for the interval

# Calculate Implementation Shortfall Day
calc_impshortfall_df['Implementation Shortfall Day'] = 0.0
calc_impshortfall_df.loc[calc_impshortfall_df['Side'] == 'B', 'Implementation Shortfall Day'] = calc_impshortfall_df['Close'] - calc_impshortfall_df['VWAP Day']
calc_impshortfall_df.loc[calc_impshortfall_df['Side'] == 'S', 'Implementation Shortfall Day'] = calc_impshortfall_df['VWAP Day'] - calc_impshortfall_df['Close']

# Calculate Implementation Shortfall Interval
calc_impshortfall_df['Implementation Shortfall Interval'] = 0.0
calc_impshortfall_df.loc[calc_impshortfall_df['Side'] == 'B', 'Implementation Shortfall Interval'] = calc_impshortfall_df['Close'] - calc_impshortfall_df['VWAP Interval']
calc_impshortfall_df.loc[calc_impshortfall_df['Side'] == 'S', 'Implementation Shortfall Interval'] = calc_impshortfall_df['VWAP Interval'] - calc_impshortfall_df['Close']

# Rank Implementation Shortfall
calc_impshortfall_df['IS Rank Day (All)'] = calc_impshortfall_df['Implementation Shortfall Day'].rank(method='dense', ascending=False).fillna(-1).astype(int)
calc_impshortfall_df['IS Rank Day (Side)'] = calc_impshortfall_df.groupby('Side')['Implementation Shortfall Day'].rank(method='dense', ascending=False).fillna(-1).astype(int)

# Reorder columns
desired_order = [
    'Ticker', 'Close', 'Side',  'Shares Interval', 'Interval Value', 'Shares Total', 'Total Value',  
    'VWAP Interval', 'VWAP Day', 'Implementation Shortfall Interval', 'Implementation Shortfall Day', 'IS Rank Day (All)', 'IS Rank Day (Side)'
]

calc_impshortfall_df = calc_impshortfall_df[desired_order]
calc_impshortfall_df
```
