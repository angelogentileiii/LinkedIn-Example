```python
import pandas as pd
from pytz import timezone

# load csv file
df = pd.read_csv('/Users/angelogentileiii/code/point-focal/russell/PointFocalFile-04_24.csv')
print(df.head())
```

```python
filteredDf = df[df['Ticker'] == 'RY']

for index, row in filteredDf.iterrows():
    total_cost = row['Price'] * row['Shares']
    print(f"Price: {row['Price']}, Shares: {row['Shares']}, Total Cost: {total_cost}")


print('Total Cost', sum(filteredDf['Shares']* filteredDf['Price']))
print('Total Shares', sum(filteredDf['Shares']))
print('VWAP Total', (sum(filteredDf['Price'] * filteredDf['Shares']))/sum(filteredDf['Shares']))
```
